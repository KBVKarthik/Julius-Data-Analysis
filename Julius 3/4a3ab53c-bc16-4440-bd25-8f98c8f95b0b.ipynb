```
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Read the dataset
df = pd.read_csv('Cars dataset.csv')

# Display basic information about the dataset
print("Dataset Info:")
print(df.info())
print("\nFirst few rows:")
print(df.head())
print("\nBasic statistics:")
print(df.describe())
```

**ai_message:**

Let me create some visualizations to better understand the data patterns.

```
# Create visualizations
plt.figure(figsize=(15, 10))

# Distribution of selling prices
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='selling_price', bins=50)
plt.title('Distribution of Selling Prices')
plt.xticks(rotation=45)

# Brand distribution
plt.subplot(2, 2, 2)
df['brand'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Car Brands')
plt.xticks(rotation=45)

# Box plot of selling price by fuel type
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='fuel', y='selling_price')
plt.title('Selling Price by Fuel Type')
plt.xticks(rotation=45)

# Scatter plot of km_driven vs selling_price
plt.subplot(2, 2, 4)
sns.scatterplot(data=df, x='km_driven', y='selling_price', alpha=0.5)
plt.title('Kilometers Driven vs Selling Price')

plt.tight_layout()
plt.show()

# Calculate average selling price by owner type
print("\nAverage Selling Price by Owner Type:")
print(df.groupby('owner')['selling_price'].mean().sort_values(ascending=False))
```

```
# Analyze top brands in detail
plt.figure(figsize=(15, 12))

# Top 10 brands by average selling price
plt.subplot(2, 2, 1)
avg_price_by_brand = df.groupby('brand')['selling_price'].mean().sort_values(ascending=False).head(10)
avg_price_by_brand.plot(kind='bar')
plt.title('Top 10 Brands by Average Selling Price')
plt.xticks(rotation=45)

# Top 10 brands by volume
plt.subplot(2, 2, 2)
df['brand'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Brands by Number of Cars')
plt.xticks(rotation=45)

# Brand distribution by fuel type
plt.subplot(2, 2, 3)
top_5_brands = df['brand'].value_counts().head(5).index
brand_fuel = df[df['brand'].isin(top_5_brands)].groupby(['brand', 'fuel']).size().unstack()
brand_fuel.plot(kind='bar', stacked=True)
plt.title('Fuel Type Distribution for Top 5 Brands')
plt.xticks(rotation=45)

# Average kilometers driven by brand
plt.subplot(2, 2, 4)
df.groupby('brand')['km_driven'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Average Kilometers Driven by Brand')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

print("\nBrand Statistics:")
brand_stats = df.groupby('brand').agg({
    'selling_price': ['mean', 'count'],
    'km_driven': 'mean'
}).sort_values(('selling_price', 'mean'), ascending=False).head(10)
print(brand_stats)
```
