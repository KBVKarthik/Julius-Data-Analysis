```
import folium
from folium import plugins
import numpy as np

# Create a base map centered on Ukraine
m = folium.Map(location=[48.3794, 31.1656], zoom_start=6)

# Sample troop movement data (in real scenario, this would come from verified sources)
# Format: [start_lat, start_lon, end_lat, end_lon, force_type]
troop_movements = [
    # Russian movements (red)
    [51.5, 32.0, 50.45, 30.52, 'Russian'],  # Movement towards Kyiv
    [52.0, 36.0, 50.0, 36.2, 'Russian'],    # Kharkiv direction
    [45.3, 36.5, 46.8, 35.0, 'Russian'],    # Southern movement
    
    # Ukrainian movements (blue)
    [49.8, 36.0, 49.1, 37.5, 'Ukrainian'],  # Eastern counter-offensive
    [48.5, 35.0, 47.8, 37.0, 'Ukrainian'],  # Southern defense
    [50.45, 30.52, 51.0, 31.5, 'Ukrainian'] # Kyiv defense
]

# Add arrows for troop movements
for movement in troop_movements:
    start_lat, start_lon, end_lat, end_lon, force = movement
    
    # Choose color based on force
    color = 'red' if force == 'Russian' else 'blue'
    
    # Create arrow coordinates
    coordinates = [[start_lat, start_lon], [end_lat, end_lon]]
    
    # Add arrow
    plugins.AntPath(
        locations=coordinates,
        color=color,
        weight=3,
        opacity=0.8,
        popup=f'{force} forces movement'
    ).add_to(m)
    
    # Add markers for start and end points
    folium.CircleMarker(
        location=[start_lat, start_lon],
        radius=5,
        color=color,
        fill=True,
        popup='Start point'
    ).add_to(m)
    
    # Add arrow endpoint
    folium.CircleMarker(
        location=[end_lat, end_lon],
        radius=5,
        color=color,
        fill=True,
        popup='End point'
    ).add_to(m)

# Add legend
m.get_root().html.add_child(folium.Element("""
    <div style="position: fixed; 
                bottom: 50px; left: 50px; width: 200px; height: 100px; 
                border:2px solid grey; z-index:9999; background-color:white;
                opacity: 0.8;">
        <p style="margin: 10px;">
            <strong>Troop Movements</strong><br>
            <span style="color: red;">→</span> Russian Forces<br>
            <span style="color: blue;">→</span> Ukrainian Forces<br>
            <small>Lines pulse to show direction</small>
        </p>
    </div>
"""))

# Save the map
m.save('ukraine_troop_movements.html')

print("Generated map showing troop movements with pulsing arrows indicating direction of movement")
```
